* **Predicting Campaign Succes**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
X = df[['age', 'job', 'contact', 'campaign', 'pdays', 'previous']]
X_encoded = pd.get_dummies(X)
y = df['y'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```

* **Identify the best time (month, day of the week) to launch a marketing campaign.**

```python
df['y'] = df['y'].apply(lambda x: 1 if x == 'yes' else 0)

monthly_success_rate = df.groupby('month')['y'].mean().reset_index()

month_order = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
monthly_success_rate['month'] = pd.Categorical(monthly_success_rate['month'], categories=month_order, ordered=True)

monthly_success_rate = monthly_success_rate.sort_values(by='month')

print(monthly_success_rate)
```

```python
weekly_success_rate = df.groupby('day_of_week')['y'].mean().reset_index()

# Sort the days of the week
day_order = ['mon', 'tue', 'wed', 'thu', 'fri']
weekly_success_rate['day_of_week'] = pd.Categorical(weekly_success_rate['day_of_week'], categories=day_order, ordered=True)

# Sort the dataframe by day_of_week
weekly_success_rate = weekly_success_rate.sort_values(by='day_of_week')

print(weekly_success_rate)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='y', data=monthly_success_rate)
plt.title('Success Rate by Month')
plt.xlabel('Month')
plt.ylabel('Success Rate')
plt.show()

# Weekly success rate plot
plt.figure(figsize=(8, 5))
sns.barplot(x='day_of_week', y='y', data=weekly_success_rate)
plt.title('Success Rate by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Success Rate')
plt.show()
```

recommendation 
* **the best month to success campaign in march and you can launch it on middel of the week (tuesday, wednesday, thursday)**

* **Predict whether a customer will subscribe to a term deposit based on other features**

```python
df.columns
```

```python
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

le = LabelEncoder()
df['y'] = le.fit_transform(df['y'])

numerical_cols = ['age', 'campaign', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']

X = df.drop('y', axis=1)
y = df['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
