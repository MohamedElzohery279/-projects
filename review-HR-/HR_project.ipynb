```python
def parse_date(date_str):
    # Convert to datetime with a default century (1900s or 2000s)
    date = pd.to_datetime(date_str, format='%d-%m-%y', dayfirst=True)
    # Adjust year based on the two-digit year rules
    if date.year < 69:
        # If year is less than 69, assume it's in 2000s
        return date.replace(year=2000 + (date.year % 100))
    else:
        # If year is 69 or more, assume it's in 1900s
        return date.replace(year=1900 + (date.year % 100))
hr_df_cleaned['DOB'] = hr_df_cleaned['DOB'].apply(parse_date)
```

2- **Understand performance, and satisfaction differences across departments and positions.**

```python
hr_df_cleaned.info()
```

```python
hr_df_cleaned['PerformanceScore'].value_counts()
```

```python
performance_mapping = {
    'Exceeds': 4,
    'Fully Meets': 3,
    'Needs Improvement': 2,
    'Poor': 1
}

hr_df_cleaned['PerformanceScore_Norm'] = hr_df_cleaned['PerformanceScore'].map(performance_mapping)

# Apply min-max normalization (scaling to 0-1 range)
hr_df_cleaned['PerformanceScore_Scaled'] = (hr_df_cleaned['PerformanceScore_Norm'] - hr_df_cleaned['PerformanceScore_Norm'].min()) / (hr_df_cleaned['PerformanceScore_Norm'].max() - hr_df_cleaned['PerformanceScore_Norm'].min())
```

```python
print(hr_df_cleaned)
```

```python
hr_df_cleaned['EmpSatisfaction'].value_counts()
```

```python
hr_df_cleaned['EmpSatisfaction_Scaled'] = (hr_df_cleaned['EmpSatisfaction'] - hr_df_cleaned['EmpSatisfaction'].min()) / (hr_df_cleaned['EmpSatisfaction'].max() - hr_df_cleaned['EmpSatisfaction'].min())

#"Exceeds" has the highest score (1.0),"Poor" has the lowest score (0.0), "Fully Meets" is scaled to approximately 0.6667, while "Needs Improvement" is scaled to 0.3333
```

```python
print(hr_df_cleaned)
```

```python
grouped_df = hr_df_cleaned.groupby(['DeptID', 'PositionID']).agg(
    AvgPerformance=('PerformanceScore_Scaled', 'mean'),
    AvgSatisfaction=('EmpSatisfaction_Scaled', 'mean')
).reset_index()

grouped_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pivot_table = grouped_df.pivot(index='DeptID', columns='PositionID', values='AvgPerformance')
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Heatmap of Performance by DeptID and PositionID")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pivot_table = grouped_df.pivot(index='DeptID', columns='PositionID', values='AvgSatisfaction')
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table, annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Heatmap of Satisfaction by DeptID and PositionID")
plt.show()
```

5- **Reduce Gender Pay Gap by 10% Over the Next 6 Months**

```python
hr_df_cleaned['New PayRate (hour)'].mean(axis=0)
```

```python
avg_pay_rate = hr_df_cleaned.groupby('GenderID')['New PayRate (hour)'].mean()
```

```python
# Calculate gender pay gap
pay_gap = (avg_pay_rate[1] - avg_pay_rate[0]) / avg_pay_rate[1] * 100
print(f"Current Gender Pay Gap: {pay_gap:.2f}%")
```

```python
gap_by_dept_position = hr_df_cleaned.groupby(['DeptID', 'PositionID', 'GenderID'])['New PayRate (hour)'].mean().unstack()

print(gap_by_dept_position)
```

```python
adjustment_factor = 1.20  

hr_df_cleaned.loc[(hr_df_cleaned['GenderID'] == 0) & (hr_df_cleaned['New PayRate (hour)'] < avg_pay_rate[1]), 'New PayRate (hour)'] *= adjustment_factor

print(hr_df_cleaned)
```

```python
avg_pay_rate_after_adjustment = hr_df_cleaned.groupby('GenderID')['New PayRate (hour)'].mean()

pay_gap_after_adjustment = (avg_pay_rate_after_adjustment[1] - avg_pay_rate_after_adjustment[0]) / avg_pay_rate_after_adjustment[1] * 100
print(f"Gender Pay Gap After Adjustments: {pay_gap_after_adjustment:.2f}%")
```

```python
def track_pay_gap(hr_df_cleaned):
    avg_pay_rate = hr_df_cleaned.groupby('GenderID')['New PayRate (hour)'].mean()
    pay_gap = (avg_pay_rate[1] - avg_pay_rate[0]) / avg_pay_rate[1] * 100
    return pay_gap

for month in range(1, 7):
    print(f"Month {month}: Gender Pay Gap = {track_pay_gap(hr_df_cleaned):.2f}%")
```

**0.34% gender pay gap is very close to equality, meaning that the average pay for male and female employees is nearly the same.**

```python
avg_tenure = hr_df_cleaned['Tenure'].mean()
print(f"Average Tenure: {avg_tenure:.2f} years")
```

9-**Track how different age groups perform and their satisfaction levels.**

```python
from datetime import datetime

hr_df_cleaned['Age'] = hr_df_cleaned['DOB'].apply(lambda x: (datetime.now() - x).days // 365)

print(hr_df_cleaned[['Employee_Name', 'DOB', 'Age']])
```

```python
hr_df_cleaned['Age'].value_counts()
```

```python
age_bins = [20, 30, 40, 50, 60, 70]  
age_labels = ['20-29', '30-39', '40-49','50-59', '60-69']

# create age groups
hr_df_cleaned['AgeGroup'] = pd.cut(hr_df_cleaned['Age'], bins=age_bins, labels=age_labels, right=False)

print(hr_df_cleaned[['Employee_Name', 'Age', 'AgeGroup']])
```

```python
age_group_analysis = hr_df_cleaned.groupby('AgeGroup').agg(
    AvgPerformance=('PerformanceScore_Norm', 'mean'),
    AvgSatisfaction=('EmpSatisfaction_Scaled', 'mean'),
    Count=('Employee_Name', 'count') 
).reset_index()

print(age_group_analysis)
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()
age_group_analysis.plot(x='AgeGroup', y='AvgPerformance', kind='bar', ax=ax1, position=0, width=0.4, color='blue', legend=False)
age_group_analysis.plot(x='AgeGroup', y='AvgSatisfaction', kind='bar', ax=ax2, position=1, width=0.4, color='green', legend=False)

ax1.set_ylabel('Average Performance', color='blue')
ax2.set_ylabel('Average Satisfaction', color='green')
plt.title('Performance and Satisfaction by Age Group')
plt.show()
```

* **Recommendation**
**Based on the age group analysis, you can create benefit packages that cater to the specific needs of different age groups. For example:**

* Younger employees (20-30): The company need to hire younger employees and they may prefer flexible work hours, educational opportunities, or health and fitness programs.
* Mid-career employees (40-50): Have higher performance with less Satisfaction,they could benefit from financial planning services, family care benefits, or career development programs.
* Older employees (50+): May be more interested in retirement plans, reduced work schedules, or mentoring opportunities.
